```python
from Scweet.scweet import scrape
from Scweet.user import get_user_information, get_users_following, get_users_followers
```

```python
data1 = scrape(words=['depression','anxiety', 'bipolar', 'mental health', 'counseling', 'self care', 'burnout', 'mental health awareness'], since="2023-03-01", until="2023-03-15", from_account = None,interval=1, 
      headless=True, display_type="Top", save_images=False, proxy = None, save_dir = 'outputs',
             resume=False, filter_replies=True, proximity=False)
```

```python
data2 = scrape(words=['depression','anxiety', 'bipolar', 'mental health', 'counseling', 'self care', 'burnout', 'mental health awareness'], since="2023-03-16", until="2023-04-01", from_account = None,interval=1, 
      headless=True, display_type="Top", save_images=False, proxy = None, save_dir = 'outputs',
             resume=False, filter_replies=True, proximity=False)
```

```python
import pandas as pd  
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

df_all_tweets= pd.concat([df1, df2], ignore_index= True)
```

```python

df_all_tweets.to_csv('project_tweets_dataset.csv')
```

```python
for x in df_all_tweets['UserName']:
    users_info= get_user_information(x, headless= True)
```

```python
users_df = pd.DataFrame(users_info, index = ["nb of following","nb of followers", "join date", 
                                             "birthdate", "location", "website", "description"]).T
users_df
```

```python
users_df.to_csv('user information.csv')
```
